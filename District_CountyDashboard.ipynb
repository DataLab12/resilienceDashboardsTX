```python
import geopandas as gpd
import pandas as pd
import panel as pn
import param
pn.extension('plotly',comms='ipywidgets')
import hvplot.pandas
import plotly.express as px
```

```python
tx_map = gpd.read_file("Data/ISDmaps/Districts2020to2021.geojson")
label_data = pd.read_csv("Data/District_Loss/DATA_Texas_District_v1.csv")
```

```python
label_data["NCES_DISTR"] = label_data["NCES_DISTR"].astype(int)
label_data = label_data.sort_values("NCES_DISTR")
```

```python
tx_map = tx_map.sort_values("NCES_DISTR")
tx_map['NCES_DISTR'] = tx_map['NCES_DISTR'].astype(int)
```

```python
merged_df = tx_map.merge(label_data, how = 'inner', on=['NCES_DISTR','DISTRICT_N'])
```

```python
merged_df
```

```python
#merged_df['% LOSS MATH'] = (merged_df['AVG_Math_2021']-merged_df['AVG_Math_2019'])/merged_df['AVG_Math_2019']
#merged_df['% LOSS READING'] = (merged_df['AVG_Reading_2021']-merged_df['AVG_Reading_2019'])/merged_df['AVG_Reading_2019']
```

```python
merged_df['Full-Time Equivalent(FTE) Teachers Difference (2021-2019)'] = (merged_df['FTE Teachers 2020-2021']-merged_df['FTE Teachers 2018-2019'])/merged_df['FTE Teachers 2018-2019']
merged_df['Sum of Full-Time Equivalent(FTE) Teachers and Staffs Difference (2021-2019)'] = (merged_df['Total Staff 2020-2021']-merged_df['Total Staff 2018-2019'])/merged_df['Total Staff 2018-2019']
merged_df['Total Number of Students Difference (2021-2019)'] = (merged_df['Total Students 2020-2021']-merged_df['Total Students 2018-2019'])/merged_df['Total Students 2018-2019']
```

```python
merged_df = merged_df.round(2)
merged_df
```

```python
merged_df.rename(columns={'% LOSS MATH':'Average Score Difference(%) in Math (2021-2019)',
                          '% LOSS READING':'Average Score Difference(%) in Reading (2021-2019)',
                          '% LOSS ALL':'Average Score Difference(%) in All (2021-2019)',
                          'Label_Math':'Label for Learning Loss in Math: (-1) Loss (0) Expected (1) Gain',
                          'Label_Reading':'Label for Learning Loss in Reading: (-1) Loss (0) Expected (1) Gain',
                          'Label_All':'Label for Learning Loss in Math and Reading: (-1) Loss (0) Expected (1) Gain'},inplace=True)
```

```python
harris_df = merged_df.loc[merged_df['County #'] == 48201]
dallas_df = merged_df.loc[merged_df['County #'] == 48113]
tarrant_df = merged_df.loc[merged_df['County #'] == 48439]
montgomery_df = merged_df.loc[merged_df['County #'] == 48339]
collin_df = merged_df.loc[merged_df['County #'] == 48085]
denton_df = merged_df.loc[merged_df['County #'] == 48121]
elpaso_df = merged_df.loc[merged_df['County #'] == 48141]
hidalgo_df = merged_df.loc[merged_df['County #'] == 48215]
fortbend_df = merged_df.loc[merged_df['County #'] == 48157]

williamson_df = merged_df.loc[merged_df['County #'] == 48491]
travis_df = merged_df.loc[merged_df['County #'] == 48453]
comal_df = merged_df.loc[merged_df['County #'] == 48091]
bastrop_df = merged_df.loc[merged_df['County #'] == 48021] 
bexar_df = merged_df.loc[merged_df['County #'] == 48029]
blanco_df = merged_df.loc[merged_df['County #'] == 48031]
burnet_df = merged_df.loc[merged_df['County #'] == 48053]
caldwell_df = merged_df.loc[merged_df['County #'] == 48055]
hays_df = merged_df.loc[merged_df['County #'] == 48209]
guadalupe_df = merged_df.loc[merged_df['County #'] == 48187]
```

```python
#Creates a county map of Texas based off of the National Center on Education Statistics DataFrame
input_list = ['Full-Time Equivalent(FTE) Teachers Difference (2021-2019)',
              'Sum of Full-Time Equivalent(FTE) Teachers and Staffs Difference (2021-2019)',
              'Total Number of Students Difference (2021-2019)','Average Score Difference(%) in Math (2021-2019)',
              'Average Score Difference(%) in Reading (2021-2019)',
              'Label for Learning Loss in Math: (-1) Loss (0) Expected (1) Gain',
              'Label for Learning Loss in Reading: (-1) Loss (0) Expected (1) Gain',
              'Label for Learning Loss in Math and Reading: (-1) Loss (0) Expected (1) Gain']
class MapDashboard(param.Parameterized):
    Attribute = param.Selector(input_list,default=input_list[0])
    @param.depends('Attribute')
    def get_map(self):
        data = harris_df[['NAME','County #','DISTRICT_N','% On Campus 10/30/20','% Tested Math G8 Diff','CARES ESSER I 20',
                    '% Black Diff','% Asian Diff','% Reduced-price Lunch Diff',self.Attribute]]
        fig = px.choropleth(data, geojson=harris_df, locations='DISTRICT_N', color= self.Attribute,
        color_continuous_scale='balance_r',
        range_color=(-0.1,0.1),
        featureidkey='properties.DISTRICT_N',
        title ='Harris County District Differences From 2019 to 2021',
        #center = {'lat':31.9686, 'lon': -99.9018},
        #zoom = 5.3, labels={'County':'County'},
        height=650, width=650,
        #mapbox_style = "white-bg",                   
        hover_data=['NAME','% On Campus 10/30/20','% Tested Math G8 Diff','CARES ESSER I 20',
                    '% Black Diff','% Asian Diff','% Reduced-price Lunch Diff',self.Attribute]
        )
        fig.update_geos(fitbounds='locations',visible=False)
        fig.update_layout(hoverlabel=dict(bgcolor='black',font_size=18,font_family="Times New Roman"),
        font=dict(family='times new roman',size=18),
        #title_x=.96,
        title_y=.97,                  
        margin={"r":10,"t":15,"l":0,"b":15})
        fig.update_coloraxes(cmid=0,colorbar_orientation='h',colorbar_y=-0.1,colorbar_title_font_family='Times New Roman',
                             colorbar_title_font_size=20,colorbar_title_side='bottom',colorbar_thickness=10)
        return fig
    def panel(self):
        return pn.Column(
            pn.Param(self, width=500, name='National Center on Education Statistics & Texas Education Agency:',sizing_mode='fixed',background='#FFFFFF'),
            self.get_map
        )
dashboard = MapDashboard(name = 'Map Dashboard')
map_component1 = dashboard.panel()
map_component1
```
